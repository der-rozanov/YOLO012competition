```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from adjustText import adjust_text
import matplotlib.font_manager as fm

# --- Настройка шрифтов ---
# Попробуем использовать Calibri, если нет - Arial
try:
    # Проверяем доступность Calibri
    calibri_path = fm.findfont(fm.FontProperties(family='Calibri'))
    plt.rcParams['font.family'] = 'Calibri'
except:
    # Fallback на Arial если Calibri нет
    plt.rcParams['font.family'] = 'Arial'
    print("Шрифт Calibri не найден, используется Arial")

# --- Загрузка данных ---
df = pd.read_csv('combined_results.csv')

# --- Настройка стиля ---
plt.style.use('ggplot')  # Чистый стиль с сеткой
sns.set_context("paper", font_scale=1.3)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12

# --- Создание графика ---
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)

# Точечный график с увеличенным размером точек
scatter = sns.scatterplot(data=df, x='Size_MB', y='metrics/mAP50-95(B)',
                         hue='Folder', s=200, palette='viridis', alpha=0.8,
                         edgecolor='black', linewidth=0.5, ax=ax)

# --- Настройка осей и заголовка ---
ax.set_xlabel('Размер модели Мб', fontweight='normal')
ax.set_ylabel('mAP50-95', fontweight='normal')
ax.set_title('Зависимость метрики mAP50-95 от размера', 
             fontweight='normal', pad=20)

# --- Умное размещение подписей ---
texts = []
for i, row in df.iterrows():
    texts.append(ax.text(row['Size_MB'], row['metrics/mAP50-95(B)'],
                       row['Folder'], 
                       fontsize=10,
                       ha='center', va='center'))

# Оптимизация расположения подписей (более агрессивные параметры)
adjust_text(texts, 
           only_move={'points':'y', 'texts':'xy'},
           arrowprops=dict(arrowstyle='-', color='gray', lw=0.5, alpha=0.5),
           expand_points=(1.5, 1.8),  # Больше "толкаем" подписи
           expand_text=(1.2, 1.5),
           force_text=(0.5, 0.8),
           force_points=(0.8, 0.8),
           lim=200)  # Максимальное количество итераций

# --- Дополнительные улучшения ---
# Легенда справа
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', 
          borderaxespad=0., title='Модели')

# Сетка
ax.grid(True, linestyle='--', alpha=0.4)

# Сохранение
plt.tight_layout()
plt.savefig('mAP50-95vs_size.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv("combined_results.csv")

# Создаем фигуру и таблицу
fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('off')  # Отключаем оси

# Рисуем таблицу (чёрный текст на белом фоне)
table = plt.table(
    cellText=df.values,
    colLabels=df.columns,
    loc='center',
    cellLoc='center',
    colColours=['#f0f0f0'] * len(df.columns)  # Серый фон заголовков (опционально)
)

# Настройка стиля
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.2)  # Масштабирование

# Сохраняем в PNG
plt.savefig("table.png", bbox_inches='tight', dpi=300, facecolor='white')
print("Таблица сохранена как 'table.png'!")
```
