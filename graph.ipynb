```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from adjustText import adjust_text
import matplotlib.font_manager as fm

# ---   ---
#   Calibri,   - Arial
try:
    #   Calibri
    calibri_path = fm.findfont(fm.FontProperties(family='Calibri'))
    plt.rcParams['font.family'] = 'Calibri'
except:
    # Fallback  Arial  Calibri 
    plt.rcParams['font.family'] = 'Arial'
    print(" Calibri  ,  Arial")

# ---   ---
df = pd.read_csv('combined_results.csv')

# ---   ---
plt.style.use('ggplot')  #    
sns.set_context("paper", font_scale=1.3)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12

# ---   ---
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)

#      
scatter = sns.scatterplot(data=df, x='Size_MB', y='metrics/mAP50-95(B)',
                         hue='Folder', s=200, palette='viridis', alpha=0.8,
                         edgecolor='black', linewidth=0.5, ax=ax)

# ---     ---
ax.set_xlabel('  ', fontweight='normal')
ax.set_ylabel('mAP50-95', fontweight='normal')
ax.set_title('  mAP50-95  ', 
             fontweight='normal', pad=20)

# ---    ---
texts = []
for i, row in df.iterrows():
    texts.append(ax.text(row['Size_MB'], row['metrics/mAP50-95(B)'],
                       row['Folder'], 
                       fontsize=10,
                       ha='center', va='center'))

#    (  )
adjust_text(texts, 
           only_move={'points':'y', 'texts':'xy'},
           arrowprops=dict(arrowstyle='-', color='gray', lw=0.5, alpha=0.5),
           expand_points=(1.5, 1.8),  #  "" 
           expand_text=(1.2, 1.5),
           force_text=(0.5, 0.8),
           force_points=(0.8, 0.8),
           lim=200)  #   

# ---   ---
#  
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', 
          borderaxespad=0., title='')

# 
ax.grid(True, linestyle='--', alpha=0.4)

# 
plt.tight_layout()
plt.savefig('mAP50-95vs_size.png', dpi=300, bbox_inches='tight')
plt.show()
```
